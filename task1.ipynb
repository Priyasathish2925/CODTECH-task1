```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Check the shape of the dataset
print(titanic.shape)

# Summary statistics for numerical features
print(titanic.describe())

# Information about missing values and data types
print(titanic.info())
```

```python
# Count missing values
print(titanic.isnull().sum())

# Drop rows with missing target (e.g., 'survived') or fill them with a reasonable value
titanic.dropna(subset=['survived'], inplace=True)
# Instead of using inplace=True, assign the modified column back to the DataFrame
titanic['age'] = titanic['age'].fillna(titanic['age'].median())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'], kde=True, bins=20)
plt.title('Age Distribution')
plt.show()

# Plot the distribution of Fare
plt.figure(figsize=(8, 6))
sns.histplot(titanic['fare'], kde=True, bins=20)
plt.title('Fare Distribution')
plt.show()

# Boxplot for Age
plt.figure(figsize=(8, 6))
sns.boxplot(x=titanic['age'])
plt.title('Age Boxplot')
plt.show()

# Boxplot for Fare
plt.figure(figsize=(8, 6))
sns.boxplot(x=titanic['fare'])
plt.title('Fare Boxplot')
plt.show()
```

```python
# Select only numeric columns
numeric_columns = titanic.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numeric_columns.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot for Age vs Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(x=titanic['age'], y=titanic['fare'], hue=titanic['survived'], palette='coolwarm')
plt.title('Age vs Fare (Survival Status)')
plt.show()
```

```python
# Countplot for 'Survived' by 'Pclass'
plt.figure(figsize=(8, 6))
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Survival Count by Pclass')
plt.show()

# Countplot for 'Survived' by 'Sex'
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Survival Count by Sex')
plt.show()
```
